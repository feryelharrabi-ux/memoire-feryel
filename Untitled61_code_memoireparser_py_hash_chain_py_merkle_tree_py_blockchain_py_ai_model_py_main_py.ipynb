```python
def parse_logs(file_path):
    logs = []
    labels = []

    with open(file_path, "r") as f:
        for line in f:
            label = 1 if line[0] == "+" else 0
            message = line[2:].strip()

            logs.append(message)
            labels.append(label)

    return logs, labels
```

```python
import hashlib

def hash_function(data):
    return hashlib.sha256(data.encode()).hexdigest()

def build_hash_chain(logs):
    hashes = []
    previous_hash = "0"

    for log in logs:
        combined = log + previous_hash
        current_hash = hash_function(combined)

        hashes.append(current_hash)
        previous_hash = current_hash

    return hashes
```

```python
%%writefile hash_chain.py
from hash_chain import hash_function

def build_merkle_tree(hashes):

    if len(hashes) == 0:
        return None

    while len(hashes) > 1:
        new_level = []

        for i in range(0, len(hashes), 2):

            if i + 1 < len(hashes):
                combined = hashes[i] + hashes[i+1]
            else:
                combined = hashes[i] + hashes[i]

            new_hash = hash_function(combined)
            new_level.append(new_hash)

        hashes = new_level

    return hashes[0]
```

```python
import hashlib
import time

blockchain = []

def create_block(index, previous_hash, merkle_root):
    block = {
        "index": index,
        "timestamp": time.time(),
        "merkle_root": merkle_root,
        "previous_hash": previous_hash
    }

    block_string = str(block).encode()
    block["hash"] = hashlib.sha256(block_string).hexdigest()

    return block

def add_block(merkle_root):

    if len(blockchain) == 0:
        previous_hash = "0"
    else:
        previous_hash = blockchain[-1]["hash"]

    block = create_block(len(blockchain)+1, previous_hash, merkle_root)
    blockchain.append(block)

    return block
```

```python
# ╔══════════════════════════════════════════════════════════════════╗
# ║     MÉMOIRE — Sécurisation des Journaux Systèmes par            ║
# ║     Hachage, Arbres de Merkle, Blockchain et IA                 ║
# ╚══════════════════════════════════════════════════════════════════╝
# Auteur   : [Ton Nom]
# Date     : 2025
# Dataset  : BGL (Blue Gene/L) ou Credit Card CSV
# ──────────────────────────────────────────────────────────────────

# ══════════════════════════════════════════════
# 0. INSTALLATION
# ══════════════════════════════════════════════
!pip install scikit-learn tensorflow pandas numpy matplotlib -q

# ══════════════════════════════════════════════
# 1. IMPORTS
# ══════════════════════════════════════════════
import os, io, re, sys, time, json, hashlib, subprocess
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

from sklearn.ensemble        import IsolationForest
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing   import StandardScaler
from sklearn.metrics         import (classification_report,
                                     confusion_matrix,
                                     f1_score, precision_score,
                                     recall_score)
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.models     import Sequential
from tensorflow.keras.layers     import (Embedding, LSTM, Dense,
                                          Dropout, Bidirectional)
from tensorflow.keras.preprocessing.text     import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks  import EarlyStopping

print("✅ Imports OK")
print(f"   TensorFlow : {tf.__version__}")
print(f"   Pandas     : {pd.__version__}")

START_GLOBAL = time.time()

# ══════════════════════════════════════════════
# 2. CHARGEMENT DES DONNÉES
# ══════════════════════════════════════════════
def load_data():
    """/content/drive/MyDrive/BGL_2k.log_structured.csv"""

    # — Monter Google Drive —
    try:
        subprocess.run(['fusermount','-u','/content/drive'],
                       capture_output=True)
        from google.colab import drive
        drive.mount('/content/drive', force_remount=True)
        drive_root = '/content/drive/MyDrive/'

        # Chercher tous les CSV disponibles
        csv_files = [f for f in os.listdir(drive_root)
                     if f.endswith('.csv')]
        print(f"📁 Fichiers CSV dans Drive : {csv_files}")

        for fname in csv_files:
            try:
                df = _read_csv_robust(drive_root + fname)
                print(f"✅ Fichier chargé : {fname}")
                return df, fname
            except Exception as e:
                print(f"⚠️  {fname} ignoré : {e}")

    except Exception as e:
        print(f"⚠️  Drive : {e}")

    # — Upload manuel —
    print("📁 Uploade ton fichier CSV :")
    from google.colab import files
    uploaded = files.upload()
    fname    = list(uploaded.keys())[0]
    df       = _read_csv_robust(io.BytesIO(uploaded[fname]))
    return df, fname

def _read_csv_robust(path):
    """Lit un CSV avec détection automatique d'encodage"""
    for enc in ['utf-8','latin-1','iso-8859-1','cp1252']:
        try:
            return pd.read_csv(path, encoding=enc)
        except Exception:
            continue
    raise ValueError(f"Impossible de lire : {path}")

# — Charger —
df_raw, filename = load_data()
print(f"\n📋 Colonnes  : {df_raw.columns.tolist()}")
print(f"📋 Shape     : {df_raw.shape}")
print(df_raw.head(3))

# ══════════════════════════════════════════════
# 3. PARSING ET PRÉPARATION
# ══════════════════════════════════════════════
def prepare_data(df):
    """
    Détecte automatiquement les colonnes texte et label.
    Supporte BGL (texte) et Credit Card (numérique).
    """
    text_candidates  = ['Content','content','Message','message',
                        'EventTemplate','Log','log','text','Text']
    label_candidates = ['Label','label','Anomaly','anomaly',
                        'Label_encoded','Class','class','Type']

    text_col  = next((c for c in text_candidates  if c in df.columns), None)
    label_col = next((c for c in label_candidates if c in df.columns), None)

    # — Fallback colonne texte —
    if text_col is None:
        candidates = {c: df[c].astype(str).str.len().mean()
                      for c in df.columns
                      if df[c].astype(str).str.len().mean() > 10}
        text_col = max(candidates, key=candidates.get) if candidates else None

    # — Fallback colonne label —
    if label_col is None:
        binary = [c for c in df.columns
                  if c != text_col and df[c].nunique() == 2]
        label_col = binary[0] if binary else None

    IS_TEXT = text_col is not None and \
              df[text_col].astype(str).str.len().mean() > 10

    print(f"\n✅ Mode détecté     : {'TEXTE (BGL)' if IS_TEXT else 'NUMÉRIQUE'}")
    print(f"   Colonne texte    : {text_col}")
    print(f"   Colonne label    : {label_col}")

    # — Échantillon équilibré (max 2000 lignes pour rapidité) —
    if label_col:
        unique_vals = df[label_col].dropna().unique()
        if set(unique_vals).issubset({0,1,'0','1'}):
            labels_raw = df[label_col].astype(int)
        else:
            normal     = str(df[label_col].value_counts().index[0]).strip()
            labels_raw = df[label_col].apply(
                lambda x: 0 if str(x).strip()==normal else 1)

        df_norm = df[labels_raw == 0].sample(
            n=min(1800, (labels_raw==0).sum()), random_state=42)
        df_anom = df[labels_raw == 1].sample(
            n=min(200,  (labels_raw==1).sum()), random_state=42)
        df_s    = pd.concat([df_norm, df_anom]).sample(
            frac=1, random_state=42).reset_index(drop=True)
    else:
        df_s      = df.sample(n=min(2000, len(df)),
                              random_state=42).reset_index(drop=True)
        label_col = '__label__'
        df_s[label_col] = 0

    # — Recalculer labels sur échantillon —
    if set(df_s[label_col].dropna().unique()).issubset({0,1,'0','1'}):
        labels = df_s[label_col].astype(int).tolist()
    else:
        normal = str(df_s[label_col].value_counts().index[0]).strip()
        labels = df_s[label_col].apply(
            lambda x: 0 if str(x).strip()==normal else 1).tolist()

    # — Logs texte —
    if IS_TEXT and text_col:
        logs = df_s[text_col].fillna("").astype(str).tolist()
    else:
        # Données numériques → convertir en pseudo-texte + garder matrice
        feat_cols = [c for c in df_s.columns if c != label_col]
        X_num     = StandardScaler().fit_transform(
                        df_s[feat_cols].fillna(0).values)
        logs      = [" ".join([f"f{i}_{v:.2f}" for i,v in enumerate(row)])
                     for row in X_num]

    print(f"   Logs préparés    : {len(logs)}")
    print(f"   Anomalies        : {sum(labels)}")
    print(f"   Normaux          : {len(labels)-sum(labels)}")

    return logs, labels, IS_TEXT, df_s

logs, labels, IS_TEXT, df_sample = prepare_data(df_raw)
y = np.array(labels)

# ══════════════════════════════════════════════
# 4. HACHAGE SHA-256 ET CHAÎNE DE HASHES
# ══════════════════════════════════════════════
print("\n" + "="*60)
print("🔐 ÉTAPE 1 — CHAÎNE DE HASHES SHA-256")
print("="*60)

def sha256(data: str) -> str:
    return hashlib.sha256(data.encode('utf-8')).hexdigest()

def build_hash_chain(logs: list) -> list:
    """
    Chaîne de hashes : H(i) = SHA256(H(i-1) + log(i))
    Garantit l'intégrité séquentielle des logs.
    """
    chain     = []
    prev_hash = "0" * 64
    ts        = time.strftime("%Y-%m-%d %H:%M:%S")

    for i, log in enumerate(logs):
        current = sha256(f"{prev_hash}{log}{i}{ts}")
        chain.append({
            "index"        : i,
            "timestamp"    : ts,
            "log"          : log[:80],   # tronquer pour lisibilité
            "previous_hash": prev_hash,
            "hash"         : current
        })
        prev_hash = current

    return chain

def verify_hash_chain(chain: list) -> bool:
    """Vérifie que chaque hash est lié au précédent"""
    for i in range(1, len(chain)):
        if chain[i]["previous_hash"] != chain[i-1]["hash"]:
            print(f"  ❌ Rupture détectée au bloc {i}")
            return False
    return True

t0          = time.time()
hash_chain  = build_hash_chain(logs)
chain_valid = verify_hash_chain(hash_chain)

print(f"  ✅ {len(hash_chain)} hashes générés en {time.time()-t0:.2f}s")
print(f"  ✅ Intégrité chaîne : {chain_valid}")
print(f"  🔑 Premier hash : {hash_chain[0]['hash'][:32]}...")
print(f"  🔑 Dernier hash : {hash_chain[-1]['hash'][:32]}...")

# Sauvegarder
with open("hash_chain.json","w") as f:
    json.dump(hash_chain[:100], f, indent=2)  # sauver 100 premiers
print("  💾 hash_chain.json sauvegardé")

# ══════════════════════════════════════════════
# 5. ARBRE DE MERKLE
# ══════════════════════════════════════════════
print("\n" + "="*60)
print("🌳 ÉTAPE 2 — ARBRE DE MERKLE")
print("="*60)

def build_merkle_tree(hashes: list) -> dict:
    """
    Construit un Merkle Tree binaire.
    Retourne la racine + tous les niveaux.
    """
    if not hashes:
        return {"root": None, "levels": [], "depth": 0}

    levels  = [hashes[:]]
    current = hashes[:]

    while len(current) > 1:
        if len(current) % 2 != 0:
            current.append(current[-1])   # dupliquer si impair
        next_lvl = [sha256(current[i] + current[i+1])
                    for i in range(0, len(current), 2)]
        levels.append(next_lvl)
        current = next_lvl

    return {
        "root"  : current[0],
        "levels": levels,
        "depth" : len(levels)
    }

def verify_merkle(logs: list, expected_root: str) -> bool:
    hashes   = [sha256(l) for l in logs]
    tree     = build_merkle_tree(hashes)
    computed = tree["root"]
    return computed == expected_root

t0           = time.time()
log_hashes   = [sha256(l) for l in logs]
merkle       = build_merkle_tree(log_hashes)
merkle_root  = merkle["root"]

print(f"  ✅ Merkle Tree construit en {time.time()-t0:.2f}s")
print(f"  ✅ Profondeur     : {merkle['depth']} niveaux")
print(f"  ✅ Feuilles       : {len(merkle['levels'][0])}")
print(f"  🌿 Racine Merkle  : {merkle_root[:32]}...")
print(f"  ✅ Vérification   : {verify_merkle(logs, merkle_root)}")

# Sauvegarder
with open("merkle_tree.json","w") as f:
    json.dump({"root": merkle_root, "depth": merkle["depth"],
               "leaves": len(logs)}, f, indent=2)
print("  💾 merkle_tree.json sauvegardé")

# ══════════════════════════════════════════════
# 6. BLOCKCHAIN
# ══════════════════════════════════════════════
print("\n" + "="*60)
print("🔗 ÉTAPE 3 — BLOCKCHAIN")
print("="*60)

class Block:
    def __init__(self, index, data, merkle_root,
                 prev_hash="0"*64, difficulty=2):
        self.index        = index
        self.timestamp    = time.strftime("%Y-%m-%d %H:%M:%S")
        self.data         = data
        self.merkle_root  = merkle_root
        self.prev_hash    = prev_hash
        self.difficulty   = difficulty
        self.nonce        = 0
        self.hash         = self._mine()

    def _compute(self) -> str:
        s = (f"{self.index}{self.timestamp}{self.data}"
             f"{self.merkle_root}{self.prev_hash}{self.nonce}")
        return sha256(s)

    def _mine(self) -> str:
        """Proof of Work simplifié"""
        target = "0" * self.difficulty
        h      = self._compute()
        while not h.startswith(target):
            self.nonce += 1
            h = self._compute()
        return h

    def to_dict(self) -> dict:
        return {
            "index"      : self.index,
            "timestamp"  : self.timestamp,
            "data"       : self.data,
            "merkle_root": self.merkle_root,
            "prev_hash"  : self.prev_hash,
            "nonce"      : self.nonce,
            "hash"       : self.hash
        }

class Blockchain:
    def __init__(self, difficulty=2):
        self.difficulty = difficulty
        self.chain      = [Block(0,"Genesis","0"*64,"0"*64,difficulty)]
        print(f"  ⛓  Bloc Genesis créé | hash={self.chain[0].hash[:20]}...")

    def add_block(self, batch_logs: list, batch_root: str):
        prev  = self.chain[-1]
        block = Block(
            index       = len(self.chain),
            data        = f"{len(batch_logs)} logs ancré",
            merkle_root = batch_root,
            prev_hash   = prev.hash,
            difficulty  = self.difficulty
        )
        self.chain.append(block)
        print(f"  ✅ Bloc #{block.index:02d} | "
              f"logs={len(batch_logs):4d} | "
              f"nonce={block.nonce:4d} | "
              f"hash={block.hash[:16]}...")
        return block

    def is_valid(self) -> bool:
        for i in range(1, len(self.chain)):
            b    = self.chain[i]
            prev = self.chain[i-1]
            if b.hash     != b._compute()  : return False
            if b.prev_hash != prev.hash    : return False
        return True

    def save(self, path="blockchain.json"):
        with open(path,"w") as f:
            json.dump([b.to_dict() for b in self.chain], f, indent=2)

t0 = time.time()
bc = Blockchain(difficulty=2)

# Découper logs en batches de 200
BATCH = 200
for i in range(0, len(logs), BATCH):
    batch      = logs[i:i+BATCH]
    batch_root = build_merkle_tree([sha256(l) for l in batch])["root"]
    bc.add_block(batch, batch_root)

bc.save("blockchain.json")
print(f"\n  ✅ Blockchain valide    : {bc.is_valid()}")
print(f"  ✅ Blocs totaux         : {len(bc.chain)}")
print(f"  ⏱  Temps               : {time.time()-t0:.2f}s")
print("  💾 blockchain.json sauvegardé")

# ══════════════════════════════════════════════
# 7. DÉTECTION IA — ISOLATION FOREST
# ══════════════════════════════════════════════
print("\n" + "="*60)
print("🤖 ÉTAPE 4A — ISOLATION FOREST (Non-supervisé)")
print("="*60)

t0          = time.time()
contamination = max(0.01, min(0.49, round(sum(labels)/len(labels), 2)))

vectorizer  = TfidfVectorizer(
    max_features  = 300,
    token_pattern = r'\S+',
    min_df        = 1,
    stop_words    = None
)
X_tfidf     = vectorizer.fit_transform(logs).toarray()

iso_model   = IsolationForest(
    n_estimators  = 100,
    contamination = contamination,
    random_state  = 42,
    n_jobs        = -1
)
iso_model.fit(X_tfidf)
iso_raw     = iso_model.predict(X_tfidf)
iso_preds   = np.where(iso_raw == -1, 1, 0)

iso_f1      = f1_score(y, iso_preds, zero_division=0)
iso_prec    = precision_score(y, iso_preds, zero_division=0)
iso_rec     = recall_score(y, iso_preds, zero_division=0)

print(f"  ⏱  Temps             : {time.time()-t0:.2f}s")
print(f"  ✅ F1-Score          : {iso_f1:.4f}")
print(f"  ✅ Précision         : {iso_prec:.4f}")
print(f"  ✅ Rappel            : {iso_rec:.4f}")
print("\n  📊 Rapport détaillé :")
print(classification_report(y, iso_preds,
      target_names=["Normal","Anomalie"], zero_division=0))

# ══════════════════════════════════════════════
# 8. DÉTECTION IA — LSTM
# ══════════════════════════════════════════════
print("\n" + "="*60)
print("🤖 ÉTAPE 4B — LSTM (Supervisé)")
print("="*60)

t0        = time.time()
MAX_WORDS = 3000
MAX_LEN   = 30
EMBED_DIM = 32
BATCH_SZ  = 64

tok = Tokenizer(num_words=MAX_WORDS, oov_token="<OOV>")
tok.fit_on_texts(logs)
X_seq = pad_sequences(
    tok.texts_to_sequences(logs),
    maxlen   = MAX_LEN,
    padding  = 'post',
    truncating='post'
)

stratify = y if len(np.unique(y)) > 1 else None
X_tr, X_te, y_tr, y_te = train_test_split(
    X_seq, y, test_size=0.2, random_state=42, stratify=stratify)

print(f"  Train : {len(X_tr)}  |  Test : {len(X_te)}")

lstm_model = Sequential([
    Embedding(MAX_WORDS, EMBED_DIM, input_length=MAX_LEN),
    Bidirectional(LSTM(32, return_sequences=True)),
    Dropout(0.3),
    LSTM(16),
    Dropout(0.2),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
], name="LSTM_AnomalyDetector")

lstm_model.compile(
    optimizer = 'adam',
    loss      = 'binary_crossentropy',
    metrics   = ['accuracy']
)
lstm_model.summary()

history = lstm_model.fit(
    X_tr, y_tr,
    epochs           = 10,
    batch_size       = BATCH_SZ,
    validation_split = 0.1,
    callbacks        = [EarlyStopping(monitor='val_loss',
                                      patience=3,
                                      restore_best_weights=True)],
    verbose = 1
)

lstm_prob  = lstm_model.predict(X_te, verbose=0).flatten()
lstm_preds = (lstm_prob >= 0.5).astype(int)

lstm_f1    = f1_score(y_te, lstm_preds, zero_division=0)
lstm_prec  = precision_score(y_te, lstm_preds, zero_division=0)
lstm_rec   = recall_score(y_te, lstm_preds, zero_division=0)

print(f"\n  ⏱  Temps             : {time.time()-t0:.2f}s")
print(f"  ✅ F1-Score          : {lstm_f1:.4f}")
print(f"  ✅ Précision         : {lstm_prec:.4f}")
print(f"  ✅ Rappel            : {lstm_rec:.4f}")
print("\n  📊 Rapport détaillé :")
print(classification_report(y_te, lstm_preds,
      target_names=["Normal","Anomalie"], zero_division=0))

# ══════════════════════════════════════════════
# 9. VISUALISATIONS
# ══════════════════════════════════════════════
print("\n" + "="*60)
print("📈 VISUALISATIONS")
print("="*60)

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle("Secure Logs — Résultats complets", fontsize=16, fontweight='bold')

# — 1. Distribution des labels —
ax = axes[0,0]
counts = [len(labels)-sum(labels), sum(labels)]
bars   = ax.bar(["Normal","Anomalie"], counts,
                color=["#2ecc71","#e74c3c"], edgecolor='white', linewidth=1.5)
ax.set_title("Distribution des logs", fontweight='bold')
ax.set_ylabel("Nombre")
for bar, count in zip(bars, counts):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+5,
            str(count), ha='center', fontweight='bold')

# — 2. Courbe Loss LSTM —
ax = axes[0,1]
ax.plot(history.history['loss'],     'b-o', label='Train Loss',  markersize=4)
ax.plot(history.history['val_loss'], 'r-o', label='Val Loss',    markersize=4)
ax.set_title("LSTM — Courbe de perte", fontweight='bold')
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend()
ax.grid(alpha=0.3)

# — 3. Courbe Accuracy LSTM —
ax = axes[0,2]
ax.plot(history.history['accuracy'],     'b-o', label='Train Acc', markersize=4)
ax.plot(history.history['val_accuracy'], 'r-o', label='Val Acc',   markersize=4)
ax.set_title("LSTM — Courbe d'accuracy", fontweight='bold')
ax.set_xlabel("Epoch")
ax.set_ylabel("Accuracy")
ax.legend()
ax.grid(alpha=0.3)

# — 4. Matrice de confusion Isolation Forest —
ax = axes[1,0]
cm = confusion_matrix(y, iso_preds)
im = ax.imshow(cm, cmap='Blues')
ax.set_title("Confusion Matrix — Isolation Forest", fontweight='bold')
ax.set_xticks([0,1]); ax.set_yticks([0,1])
ax.set_xticklabels(["Normal","Anomalie"])
ax.set_yticklabels(["Normal","Anomalie"])
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm[i,j], ha='center', va='center',
                fontsize=14, fontweight='bold',
                color='white' if cm[i,j] > cm.max()/2 else 'black')

# — 5. Matrice de confusion LSTM —
ax = axes[1,1]
cm2 = confusion_matrix(y_te, lstm_preds)
ax.imshow(cm2, cmap='Oranges')
ax.set_title("Confusion Matrix — LSTM", fontweight='bold')
ax.set_xticks([0,1]); ax.set_yticks([0,1])
ax.set_xticklabels(["Normal","Anomalie"])
ax.set_yticklabels(["Normal","Anomalie"])
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm2[i,j], ha='center', va='center',
                fontsize=14, fontweight='bold',
                color='white' if cm2[i,j] > cm2.max()/2 else 'black')

# — 6. Comparaison F1 / Précision / Rappel —
ax    = axes[1,2]
metr  = ['F1-Score','Précision','Rappel']
iso_v = [iso_f1, iso_prec, iso_rec]
lst_v = [lstm_f1, lstm_prec, lstm_rec]
x     = np.arange(len(metr))
w     = 0.35
ax.bar(x-w/2, iso_v,  w, label='Isolation Forest', color='#3498db', alpha=0.85)
ax.bar(x+w/2, lst_v,  w, label='LSTM',             color='#e67e22', alpha=0.85)
ax.set_title("Comparaison des modèles", fontweight='bold')
ax.set_xticks(x); ax.set_xticklabels(metr)
ax.set_ylim(0, 1.1)
ax.legend()
ax.grid(axis='y', alpha=0.3)
for xi, (iv, lv) in enumerate(zip(iso_v, lst_v)):
    ax.text(xi-w/2, iv+0.02, f"{iv:.2f}", ha='center', fontsize=9)
    ax.text(xi+w/2, lv+0.02, f"{lv:.2f}", ha='center', fontsize=9)

plt.tight_layout()
plt.savefig("resultats.png", dpi=150, bbox_inches='tight')
plt.show()
print("  💾 resultats.png sauvegardé")

# ══════════════════════════════════════════════
# 10. RAPPORT FINAL
# ══════════════════════════════════════════════
print("\n" + "╔"+"═"*58+"╗")
print("║" + "   RAPPORT FINAL — SECURE LOGS".center(58) + "║")
print("╠" + "═"*58 + "╣")

total_time = time.time() - START_GLOBAL

lines = [
    ("Fichier analysé",         filename),
    ("Logs totaux",             str(len(logs))),
    ("Anomalies réelles",       f"{sum(labels)} ({100*sum(labels)/len(labels):.1f}%)"),
    ("Normaux",                 f"{len(labels)-sum(labels)}"),
    ("─── Intégrité ───",       ""),
    ("Hashes SHA-256",          str(len(hash_chain))),
    ("Chaîne valide",           str(chain_valid)),
    ("Merkle root",             merkle_root[:24]+"..."),
    ("Blocs blockchain",        str(len(bc.chain))),
    ("Blockchain valide",       str(bc.is_valid())),
    ("─── IA ───",              ""),
    ("Isolation Forest F1",     f"{iso_f1:.4f}"),
    ("Isolation Forest Préc.",  f"{iso_prec:.4f}"),
    ("Isolation Forest Rappel", f"{iso_rec:.4f}"),
    ("LSTM F1",                 f"{lstm_f1:.4f}"),
    ("LSTM Précision",          f"{lstm_prec:.4f}"),
    ("LSTM Rappel",             f"{lstm_rec:.4f}"),
    ("Meilleur modèle",         "LSTM 🥇" if lstm_f1 > iso_f1
                                 else "Isolation Forest 🥇"),
    ("─── Temps ───",           ""),
    ("Temps total",             f"{total_time:.2f}s"),
]

for k, v in lines:
    if v == "":
        print("║" + f"  {k}".ljust(58) + "║")
    else:
        print("║" + f"  {k:<30} {v}".ljust(58) + "║")

print("╚" + "═"*58 + "╝")

# ══════════════════════════════════════════════
# 11. SAUVEGARDER TOUT SUR GOOGLE DRIVE
# ══════════════════════════════════════════════
try:
    drive_out = '/content/drive/MyDrive/secure-logs-results/'
    os.makedirs(drive_out, exist_ok=True)

    files_to_save = [
        "hash_chain.json",
        "merkle_tree.json",
        "blockchain.json",
        "resultats.png"
    ]
    for f in files_to_save:
        if os.path.exists(f):
            import shutil
            shutil.copy(f, drive_out + f)

    # Sauvegarder le modèle LSTM
    lstm_model.save(drive_out + "lstm_model.h5")

    print(f"\n✅ Tous les fichiers sauvegardés dans Drive :")
    print(f"   {drive_out}")
    for f in os.listdir(drive_out):
        print(f"   📄 {f}")

except Exception as e:
    print(f"⚠️  Sauvegarde Drive : {e}")

print("\n🎓 Projet terminé avec succès !")
```
